#### Handling missing values 

```python
data_mean = data.groupby(['Education']).mean()['Income'].reset_index()
data_mean
```

```python
df_g = data.copy()
df_g.head()
```

```python
df_g.groupby(['Education']).mean()['Income']
```

```python
df_g.columns
```

```python
df_g[df_g['Income'].isna()]
```

```python
df_g['Income'] = df_g.groupby('Education')['Income'].transform(lambda x : x.fillna(np.mean(x)))
```
