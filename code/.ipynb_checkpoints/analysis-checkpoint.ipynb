##### Dealing with missing values

###### Strategy 1: Impute mean

```python
data_mean = data.copy()
income_mean = data_mean['Income'].mean()
print(income_mean)
```

```python
data_mean['Income'].fillna(income_mean, inplace = True)
data_mean['Income']
```
