### Important Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set(color_codes = True)
sns.set(rc={'figure.figsize': (13,10.1)})
pd.options.display.float_format = "{:.3f}".format
sns.set_theme(style="whitegrid")
```

### Loading Dataset

```python
data_set = pd.read_excel('../data/marketing_campaign.xlsx', index_col = 0)
data_set.head()
```

### Dealing with duplicate values

```python
# How many Duplicate rows are here??

dup_rows= data_set[data_set.duplicated()]
print(dup_rows.shape)
```

```python
# Dropping Duplicates

data_set = data_set.drop_duplicates()
print(data_set[data_set.duplicated()].shape)
```

```python
# Columns

print(data_set.columns)
```

### Dealing with missing values

##### Checking for missing values


```python
# Missing value percentage
data_set.isna().sum()/len(data_set)
```

```python
# Missing value Datas

data_set[data_set['Income'].isna()]
```

##### Replacing missing values

```python
# Theoretically, 25% to 30% is the maximum missing values are allowed
# This just contains only 1% missing value
# But i will replace those missing values with grouped data values[MEAN] of Income column
```

```python
df = data_set.copy()
df.head()
```

```python
df['Income'].isna().sum()
```

```python
# Filling null values with grouped Mean

df['Income'] = df.groupby('Education')['Income'].transform(lambda group : group.fillna(np.mean(group)))
```

```python
df['Income'].isna().sum()
```

```python
# No missing values
```

#### Clearing up some columns for analysis

```python
# Creating new columns based on existing columns

df['Age'] = 2014 - df['Year_Birth']
df['Child_in_home'] = df['Kidhome'] + df['Teenhome']
df['Camp_accepted'] = df['AcceptedCmp3'] + df['AcceptedCmp4'] + df['AcceptedCmp5'] + df['AcceptedCmp1'] + df['AcceptedCmp2']
df['Item Purchased'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
```

```python
# Dropping unnecessery columns

df.drop(['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2', 'Kidhome', 'Teenhome', 'MntWines', 'MntFruits','MntMeatProducts', 'MntFishProducts', 'MntSweetProducts','MntGoldProds','Year_Birth'], axis = 1, inplace = True)
```

```python
df.columns
```

```python
# Some more column clearing
# Minimizing variability
df['Marital_Status'].replace({'Divorced':'Alone','Single':'Alone','Married':'In couple','Together':'In couple','Absurd':'Alone','Widow':'Alone','YOLO':'Alone'}, inplace = True)
```

##### Handling Outlires

```python
# First we will find the Inter-Quantile Range (IQR) of my Dataset

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# Shape of the current Dataset
df.shape
```

```python
# Removing outlires

df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# Shape of the Dataset after removing outlires

df.shape
```
