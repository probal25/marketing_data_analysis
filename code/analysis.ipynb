##### Dealing with missing values

###### Strategy 1: Impute mean

```python
data_mean = data.copy()
income_mean = data_mean['Income'].mean()
print(income_mean)
```

```python
data_mean['Income'].fillna(income_mean, inplace = True)
data_mean['Income']
```

```python
Q1 = data_mean.quantile(0.25)
Q3 = data_mean.quantile(.75)

IQR = Q3 - Q1
print(IQR)
```

```python
data_mean = data_mean[~((data_mean < (Q1 - 1.5 * IQR)) | (data_mean > (Q3 + 1.5 * IQR))).any(axis=1)]
data_mean.shape
```
